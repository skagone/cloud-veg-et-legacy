```python
import os
import numpy as np
import rasterio
from matplotlib import pyplot as plt
```

```python
var='model_outputs/etasw'
yr = 2014
mon = 2
file = '{}_{}{:02d}.tif'.format(var, yr, mon)
ds = rasterio.open(file)
band1 = ds.read(1)
plt.figure(figsize=(4,8))
plt.imshow(band1, vmin=0, vmax=50, cmap='viridis')
plt.show()
```

```python

var='model_outputs/et'
yr = 2014

for i in range(9):
    day = i+1

    file = '{}_{}{:03d}.tif'.format(var, yr, day)
    file = '{}_{}{:02d}.tif'.format(var, yr, 7)

    ds = rasterio.open(file)

    band1 = ds.read(1)

    plt.figure(figsize=(10,12))
    plt.imshow(band1, vmin=50, vmax=200, cmap='viridis')
    plt.show()

#from rasterio.plot import show
#show(ds)

#test_ds = rasterio.open('test_tif.tif')
#show(test_ds, cmap='viridis', adjust='linear')
```
