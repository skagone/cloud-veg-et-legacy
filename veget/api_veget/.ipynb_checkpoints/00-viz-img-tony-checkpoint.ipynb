```python
import rasterio
import matplotlib.pyplot as plt

ps3='/vsis3/dev-et-data/tile_40N-90E/etasw_2013006.tif'
with rasterio.open(ps3, 'r') as src:    
     plt.imshow(src.read(1), vmin=0, vmax=5, cmap='pink')
     plt.show()
    
     plt.imshow(src.read(1), vmin=0, vmax=5, cmap='Accent_r')
     plt.show()
```

```python
import rasterio
import matplotlib.pyplot as plt

cmaps = ['gnuplot', 'gnuplot2', 'gnuplot2_r', 'gnuplot_r', 'gray', 'gray_r', 'hot', 'hot_r', 'hsv', 'hsv_r']

ps3='/vsis3/dev-et-data/tile_40N-90E/etasw_2013006.tif'
with rasterio.open(ps3, 'r') as src:    
     plt.imshow(src.read(1), vmin=0, vmax=5, cmap='pink')
     plt.show()
    
     for cmap in cmaps:
        plt.imshow(src.read(1), vmin=0, vmax=5, cmap=cmap)
        plt.show()
     
```
